# 5 - Rendimento da madeira serrada (R)

![Diagrama](4-5-diagrama.png)

$$
R = \frac{V_{mad.serrada}}{V_{tora}} \times 100
$$

onde:

* $R \rightarrow$ *rendimento em %*;
* $V_{mad.serrada} \rightarrow$ *volume de madeira serrada, em $m^3$ ($V_{real}$ ou $V_{nominal}$)*; e
* $V_{tora} \rightarrow$ *volume da tora, em $m^3$ (com ou sem casca)*.

## 5.1 - Cálculo de perdas no processo de desdobro

$$
\%Perdas_{toras S.C.} = \left( \frac{V_{serragem} + V_{costaneiras,refugos}}{V_{tora S.C.}} \right) \times 100
$$

$$
\%Perdas_{toras C.C.} = \left( \frac{V_{serragem} + V_{costaneiras,refugos}}{V_{tora C.C.}} \right) \times 100
$$

$$
V_{costaneira,refugos} = V_{tora S.C.} - \left( V_{mad.settada} + V_{serragem} \right)
$$

$$
V_{costaneira,refugos} = V_{tora C.C.} - \left( V_{mad.settada} + V_{serragem} + V_{cascas} \right)
$$

$$
V_{casca} = V_{tora S.C.} - V_{tora C.C.}
$$

onde:

* $S.C. \rightarrow$ *Sem Casca*; e
* $C.C. \rightarrow$ *Com Casca*.

### Exemplo

Calcular o rendimento real da madeira serrada do notebook 2. Considere o volume real das tábuas do bloco e laterais.

```python
# Dados do cálculo do volume máximo obtido no notebook 1
# ------------------------------------------------------------------------------------------------------
D1 = 35        # centímetros
D2 = 30        # centímetros
c = 2.75       # metros
b = 3          # milímetros

alpha = 25.09  # constante obtida acima
# ------------------------------------------------------------------------------------------------------

# Dados do cálculo do volume máximo obtido no notebook 2
# ------------------------------------------------------------------------------------------------------
espessura_normalizada = 25                             # milímetros
sobredimensao_espessura = espessura_normalizada * 0.05 # 5% de espessura_normalizada
largura_corte_serra = 3                                # milímetros
# ------------------------------------------------------------------------------------------------------
```

```python
# Solução do cálculo do volume máximo obtido no notebook 1
# ------------------------------------------------------------------------------------------------------
import math

_round = 4

a = round((D2 / math.sqrt(2)) / 100,_round)
volume_bloco_maximo = round(a**2 * c,_round)
volume_tora = round((math.pi * (D1**2 / 100 + D2**2 / 100) / 8) * c / 100,_round)
rendimento = round((volume_bloco_maximo / volume_tora) * 100,_round)

R = round(D2 / 2 / 100,_round)
K = round(a / 2,_round)
x = round((R * math.cos(math.radians(alpha)) - K) * 1000,_round)
x_linha = x - b

y = round(2 * R * math.sin(math.radians(alpha)) * 1000,_round)

volume_4_pecas_laterais = round(4 * (x_linha / 1000 * y / 1000) * c,_round)

rendimento_maximo = round(((volume_bloco_maximo + volume_4_pecas_laterais) / volume_tora) * 100,_round)
# ------------------------------------------------------------------------------------------------------

# Solução do cálculo do volume máximo obtido no notebook 2
# ------------------------------------------------------------------------------------------------------
espessura_corte = espessura_normalizada + sobredimensao_espessura + largura_corte_serra
numero_tabuas = round(a * 1000 / espessura_corte)

espessura = espessura_normalizada * 1.05

largura = a

volume_pecas_bloco = round(numero_tabuas * (espessura / 1000 * largura * c),_round)

rendimento_maximo_real = round(((volume_4_pecas_laterais + volume_pecas_bloco) / volume_tora) * 100)
# ------------------------------------------------------------------------------------------------------

# Solução do cálculo do volume máximo obtido no notebook 2
# ------------------------------------------------------------------------------------------------------
_round = 6

volume_corte_1 = round(2 * (a * b * c) / 1000,_round)
volume_corte_2 = round(2 * (a * b * c) / 1000,_round)
volume_corte_3 = round(7 * (a * b * c) / 1000,_round)
volume_corte_4 = round(4 * (y / 1000 * b * c) / 1000,_round)
volume_corte_5 = round(8 * (x_linha / 1000 * b * c) / 1000,_round)

volume_total_serragem = volume_corte_1 + volume_corte_2 + volume_corte_3 + volume_corte_4 + volume_corte_5

perda_serragem = round((volume_total_serragem / volume_tora) * 100)
# ------------------------------------------------------------------------------------------------------

# Solução deste exemplo
_round = 6
volume_madeira_serrada = volume_4_pecas_laterais + volume_pecas_bloco
volume_costaneiras_refugos = round(volume_tora - (volume_madeira_serrada + volume_total_serragem),_round)
perdas_toras_sem_casca = round(((volume_total_serragem + volume_costaneiras_refugos) / volume_tora) * 100)
razao = round((volume_madeira_serrada / volume_tora) * 100)
```

##### $V_{costaneira,refugos} = $ {{ volume_tora }} - ( {{ volume_madeira_serrada }} + {{ volume_total_serragem}} )
$V_{costaneira,refugos} = $ {{ volume_costaneiras_refugos }} $_{m^3}$

##### $\%Perda_{serragem} = $ ( {{ volume_total_serragem }} / {{ volume_tora }} ) $\times 100$
$\%Perda_{serragem} = $ {{ perda_serragem }} %

##### $\%Perdas_{toras S.C.} = \left( \frac{V_{serragem} + V_{costaneira,refugos}}{V_{tora S.C.}} \right)$

$\%Perdas_{toras S.C.} = $ (( {{ volume_total_serragem }} $_{m^3}$ + {{ volume_costaneiras_refugos }} $_{m^3}$ ) / {{ volume_tora }} ) $\times 100$

$\%Perdas_{toras S.C.} = $ {{ perdas_toras_sem_casca }} %

##### $R = \frac{V_{mad.serrada}}{V_{tora}} \times 100 (\%)$

$R = $( {{ volume_madeira_serrada }} / {{ volume_tora }}) * 100
